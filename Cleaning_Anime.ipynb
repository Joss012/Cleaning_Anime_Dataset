# Data Cleaning

- `drop` columns that will not be useful

### The "Premiered" column has a lot of missing values but can be fixed. 
-  PLAN:
    - Extract the Month and Year from the "Aired" column
    - Convert the Month into a season
    - Create two new columns with the season and year the anime premiered
    - Drop existing "Premiered" column

```python
results = df.query('Premiered.str.contains("Unknown")')
results.head()
```

### Extracting the Month from the "Aired" column

```python
aired_month = df['Aired'].str.slice(0,3)
aired_month.head()
```

### Extracting the Year from the "Aired" column

```python
aired_year = df['Aired'].str.split(',').str[1].str.slice(0,5)
aired_year.head()
```

### Helper method to convert the extracted months into the correct anime seasons

```python
def seasons(month): 

   if(month == ('Jan')):  
    return 'Winter'
   elif(month == ('Feb')):
    return 'Winter'
   elif(month == ('Mar')):
    return 'Winter'
   elif(month == ('Apr')):
    return 'Spring'
   elif(month == ('May')):
    return 'Spring'
   elif(month == ('Jun')):
    return 'Spring'
   elif(month == ('Jul')):
    return 'Summer'
   elif(month == ('Aug')):
    return 'Summer'
   elif(month == ('Sep')):
    return 'Summer'
   elif(month == ('Oct')):
    return 'Fall'
   elif(month == ('Nov')):
    return 'Fall'
   elif(month == ('Dec')):
    return 'Fall'
   else: 
    return 'Unknown'
```

### Converting the months into the correct anime seasons

```python
aired_season = aired_month.apply(seasons)
aired_season.head()
```

### Creating the two new "Season_Premiered" and "Year_Premiered" columns and Dropping the existing "Premiered" column

```python
df.drop(columns={'Aired', 'Premiered'}, inplace=True)
df.insert(5, "Season_Premiered", aired_season)
df.insert(6, "Year_Premiered", aired_year)
df.head()
```

### Filling in NaN values in the "Year_Premiered" column with "Unknown" so they can be easily dropped later 

```python
df['Year_Premiered'] = df['Year_Premiered'].fillna("Unknown")
```

### After the cleaning of the "Premiered" column there are a lot less "Unknown" values

```python
df.map(lambda x: "Unknown" in str(x)).sum()
```

### Checking to make sure that no rows were lost during the cleaning and that only values were replaced

```python
df.shape
```

### The "Genres" column has too many genres and will be trimmed to only the main genre

#### Before trim

```python
df.head()
```

#### After trim

```python
genre = df['Genres'].str.split(',').str[0]
df.drop(columns='Genres', inplace=True)
df.insert(2, "Genre", genre)
df.head()
```

### The "Rating" column can be trimmed to only the rating itself without the description 

#### Before trim

```python
df.head()
```

#### After trim

```python
rating = df['Rating'].str.split(' ').str[0]
df.drop(columns='Rating', inplace=True)
df.insert(7, "Rating", rating)
df.head(10)
```

### Dropping the rows that still have "Unknown" values after the cleaning 

```python
df = df[(df != 'Unknown').all(axis=1)]
# resetting the index so it is still in order after dropping values 
df.reset_index(inplace=True, drop=True)
```

```python
df.map(lambda x: "Unknown" in str(x)).sum()
```

### The one "Name" row that has "Unknown" in it is not empty. "Unknown" is apart of its title

```python
df.query('Name.str.contains("Unknown")')
```

### Converting "Year_Premiered", "Episodes", and "Ranked" columns to type int after all of the "Unknown" columns were dropped

```python
df['Year_Premiered'] = df['Year_Premiered'].astype(int)
df['Episodes'] = df['Episodes'].astype(int)
df['Ranked'] = df['Ranked'].astype(int)
```
